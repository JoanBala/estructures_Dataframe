###### Departure Delays

```python
df_Dep_Diff_time = df_flights_copia[df_flights_copia.isnull().any(axis=1)][['UniqueCarrier']]
#df_Dep_Diff_time
```

```python
df_Dep_Diff_time["Delayed"] = df_flights_copia["DepTime"] - df_flights_copia["CRSDepTime"] 
df_Dep_Diff_time

mask = df_Dep_Diff_time["Delayed"] > 15

df_Dep_Diff_time = df_Dep_Diff_time[mask]
print("Vols amb retard superior a 15 minuts a la Sortida")
df_Dep_Diff_time
```

```python
print("Número de vols amb retard superior a 15min agrupats per companyies:")
df_vols_retard_Dep = df_Dep_Diff_time['UniqueCarrier'].value_counts().sort_values()
df_vols_retard_Dep
```

```python
#vols per companyia
print("Número de vols per companyia:")
df_Num_of_flights = df_flights_copia['UniqueCarrier'].value_counts().sort_values()
df_Num_of_flights
```

```python
# % Percentatge de vols amb retard >15min respecte del total, per companyia.
df_Num_of_flights_PT_Dep = round(df_vols_retard_Dep / df_Num_of_flights * 100, 2).sort_values()
df_Num_of_flights_PT_Dep
```

###### Arrival Delays

```python
df_Arr_Diff_time = df_flights_copia[df_flights_copia.isnull().any(axis=1)][['UniqueCarrier']]
#df_Arr_Diff_time
```

```python
df_Arr_Diff_time["Delayed"] = df_flights_copia["ArrTime"] - df_flights_copia["CRSArrTime"] 
df_Arr_Diff_time

mask = df_Arr_Diff_time["Delayed"] > 15

df_Arr_Diff_time = df_Arr_Diff_time[mask]
print("Vols amb retard superior a 15 minuts a l'Arribada")
df_Arr_Diff_time
```

```python
print("Número de vols amb retard superior a 15min agrupats per companyies:")
df_vols_retard_Arr = df_Arr_Diff_time['UniqueCarrier'].value_counts().sort_values()
df_vols_retard_Arr
```

```python
# % Percentatge de vols amb retard >15min respecte del total, per companyia.
df_Num_of_flights_PT_Arr = round(df_vols_retard_Arr / df_Num_of_flights * 100, 2).sort_values()
df_Num_of_flights_PT_Arr
```

```python
DF_retards_companyies = pd.DataFrame({'Vols_totals':df_Num_of_flights,
                                      '%_Perdut_Sortides': df_Num_of_flights_PT_Dep,
                                      'Num_vols_retard_Sortides':df_vols_retard_Dep,
                                      '%_Perdut_Arribades': df_Num_of_flights_PT_Arr,
                                      'Num_vols_retard_Arribades':df_vols_retard_Arr})
DF_retards_companyies.index.name="Companyies"
DF_retards_companyies            #.sort_values(by=['Vols_totals'], axis=0, ascending=True)
```

```python
print(f"La companyia amb més retards de Sortida és: ",DF_retards_companyies['%_Perdut_Sortides'].idxmax(axis=1))
```

```python
print(f"La companyia amb més retards d'Arribada és: ",DF_retards_companyies['%_Perdut_Arribades'].idxmax(axis=1))
```
